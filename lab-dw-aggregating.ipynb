```python
# 1. Create a new DataFrame that only includes customers who have a total_claim_amount greater than $1,000 and have a response of "Yes" to the last marketing campaign.

import pandas as pd


data = pd.read_csv('https://raw.githubusercontent.com/data-bootcamp-v4/data/main/marketing_customer_analysis.csv')

data
```

```python
filtered_data = data[(data['Total Claim Amount'] > 1000) & (data['Response'] == 'Yes')]   # Filter the data

new_df = pd.DataFrame(filtered_data)  # Create a new DataFrame
new_df
```

```python
# 2. Using the original Dataframe, analyze the average total_claim_amount by each policy type and gender for customers who have responded "Yes" to the last marketing campaign. 
# Write your conclusions.
```

```python

g_data = data[data['Response'] == 'Yes'].groupby(['Policy Type', 'Gender'])['Total Claim Amount'].mean().reset_index()

g_data# Group by policy type and gender for customers who responded "Yes"
```

```python
g_data.plot(kind='bar', x='Policy Type', y='Total Claim Amount')
```

```python
""" 
Looking at the result we could say that all fields are very similar.
Corporate Auto has a little more F
Personal Auto has a little more M
and Special Auto has a little more F
"""
```

```python
#3. Analyze the total number of customers who have policies in each state, and then filter the results to only include states where there are more than 500 customers.
```

```python

st_customers = data.groupby('State')['Customer'].nunique().reset_index(name='total_customers')
st_customers


```

```python
states_plus_500_customers = st_customers[st_customers['total_customers'] > 500]  # Filter states with more than 500 customers
states_plus_500_customers
```

```python
# to test for more then 900 customeres 
states_plus_900_customers = st_customers[st_customers['total_customers'] > 900]  # Filter states with more than 500 customers
states_plus_900_customers
```

```python
# 4. Find the maximum, minimum, and median customer lifetime value by education level and gender. Write your conclusions.
```

```python
clf_stats = data.groupby(['Education', 'Gender'])['Customer Lifetime Value'].agg(['max', 'min', 'median']).round(2) # I extra Rounded to 2 decimal numbers


print(clf_stats )
```

```python
import matplotlib.pyplot as plt

# Assuming 'lifetime_stats' is the dataframe containing the maximum, minimum, and median customer lifetime value by education level and gender
# Replace 'lifetime_stats' with the actual variable name of your dataframe

# Plotting the results
fig, axes = plt.subplots(nrows=2, ncols=2, figsize=(12, 10))

# Maximum lifetime value
clf_stats['max'].unstack().plot(kind='bar', ax=axes[0, 0], title='Maximum Lifetime Value by Education Level and Gender')

# Minimum lifetime value
clf_stats['min'].unstack().plot(kind='bar', ax=axes[0, 1], title='Minimum Lifetime Value by Education Level and Gender')

# Median lifetime value
clf_stats['median'].unstack().plot(kind='bar', ax=axes[1, 0], title='Median Lifetime Value by Education Level and Gender')

plt.tight_layout()
plt.show()
```

```python
""" 
In Maximum Lifetime Value by Education Level and Gender we can see that the highest number of people are in the High School or Below and Male
In Minimum Lifetime Value by Education Level and Gender we can see that the highest number of people are in the Doctor and Female and very close to the value of female master
in Median Lifetime Value by Education Level and Gender we can see that the highest number of people are in the High School or Below and Male
    """ 
```
